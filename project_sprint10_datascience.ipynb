### 4 - Fun√ß√£o para c√°lculo de lucro:

````python
def profit_model(data, sample_size=500, top_x=200, valor_produto=4500, invest_poco=100000000/200):
#Subdivindo os dados em features e target:
    features = data.drop(['id','product'], axis=1)
    target = data['product']

#Divis√£o entre conjuntos de treinamento e valida√ß√£o:
    features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=0.25, random_state=12345)

# SAMPLE apenas no conjunto de valida√ß√£o (se necess√°rio)
    sample_indices = features_valid.sample(n=sample_size, random_state=12345).index
    features_valid = features_valid.loc[sample_indices]
    target_valid = target_valid.loc[sample_indices]

#Treinamento do modelo e predi√ß√µes do conjunto de valida√ß√£o, e por fim transformando o objeto array em um Series:
    model = LinearRegression()
    model.fit(features_train, target_train)
    predicted_valid = model.predict(features_valid)
    predicted_valid = pd.Series(predicted_valid, index=target_valid.index)

#C√°lculo de lucro para os 200 melhores po√ßos de acordo com a previs√£o:
    best_x = predicted_valid.sort_values(ascending=False).head(top_x)
    profit = best_x.sum() * valor_produto - best_x.count() * valor_produto
    print(f"O lucro estimado dos {top_x} melhores po√ßos encontrados no modelo √© igual a ${profit}")

#Armazenando valores de volume reais dos po√ßos com melhores previs√µes pelo modelo treinado:
    best_x_target = target_valid.loc[best_x.index]
    
#C√°lculo do REQM para regress√£o linear:
    mse = mean_squared_error(best_x, best_x_target)
    print('Regress√£o Linear')
    print('EQM =', mse)
    print('REQM =', mse ** 0.5)
    print('Volume m√©dio previsto =', best_x.mean())

    return best_x
````

Esta regi√£o n√£o apenas promoveu a maior m√©dia estimada de volume de petr√≥leo nos 200 maiores po√ßos, como tamb√©m apresentou melhor REQM do que as demais.

### 5 - Bootstrapping para c√°lculo de lucro:

````python
def bootstrapping_profit(data, sample_size=500, samples=1000, wells=200):
    state = np.random.RandomState(54321)
    profits = []
    
# Preparar dados
    features = data.drop(['id','product'], axis=1)
    target = data['product']
    
# Dividir dados
    features_train, features_valid, target_train, target_valid = train_test_split(
        features, target, test_size=0.25, random_state=12345)

#Treinamento do modelo e predi√ß√µes do conjunto de valida√ß√£o, e por fim transformando o objeto array em um Series:
    model = LinearRegression()
    model.fit(features_train, target_train)
    predicted_valid = model.predict(features_valid)
    predicted_valid = pd.DataFrame({'Predito':predicted_valid, 'Real':target_valid.values}, index=target_valid.index)
    
    for i in range(samples):
        # Reamostragem dos 200 melhores po√ßos
        subsample = predicted_valid.sample(n=sample_size, replace=True, random_state=state)
        best_wells = subsample.sort_values(by='Predito', ascending=False).head(wells)
        
        # Calcular lucro da amostra
        profit = best_wells['Real'].sum() * valor_produto - 100000000
        profits.append(profit)
        
        # Atualizar o estado para pr√≥xima itera√ß√£o
        state = np.random.RandomState(state.randint(0, 10000))
    return pd.Series(profits)
````

````python
#C√°lculo de lucro total:

profit_0 = bootstrapping_profit(data0)
print(f"O lucro estimado da Regi√£o 01, ap√≥s aplica√ß√£o da t√©cnica de Bootstrapping, √© igual a ${profit_0.sum()}")
profit_1 = bootstrapping_profit(data1)
print(f"O lucro estimado da Regi√£o 02, ap√≥s aplica√ß√£o da t√©cnica de Bootstrapping, √© igual a ${profit_1.sum()}")
profit_2 = bootstrapping_profit(data2)
print(f"O lucro estimado da Regi√£o 03, ap√≥s aplica√ß√£o da t√©cnica de Bootstrapping, √© igual a ${profit_2.sum()}")
````

````python
#C√°lculo de m√©dio por po√ßo:

print(f"O lucro m√©dio da Regi√£o 01, ap√≥s aplica√ß√£o da t√©cnica de Bootstrapping, √© igual a ${profit_0.mean()}")
print(f"O lucro m√©dio da Regi√£o 02, ap√≥s aplica√ß√£o da t√©cnica de Bootstrapping, √© igual a ${profit_1.mean()}")
print(f"O lucro m√©dio da Regi√£o 03, ap√≥s aplica√ß√£o da t√©cnica de Bootstrapping, √© igual a ${profit_2.mean()}")
````
